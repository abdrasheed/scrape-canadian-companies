### Scrape Jobs Data From Wuzzaf site

1- Import Necessary Libraries

```python
# Get requests links 
import requests

# Convert text to elements 
from bs4 import BeautifulSoup

# Create and store Data to csv file
import csv

# Sleep some time
import time

# Pandas Data Frame
import pandas as pd

import re
```

2- Start to scrape data we need. <br><strong>[company_name | city | Number_of_employees | projects_pudget | url]</strong>

2-1- Scrape the feature data and store it in the CSV file

```python
with open('Data Scraped/canada_companies.csv','w') as canada:
    wr = csv.writer(canada)
    wr.writerow(['company_name','city','Number_of_employees','projects_pudget','url'])
    
    for i in range(1,5):
        if i != 1:
            # Concatinate counter with pages
            url = 'https://themanifest.com/ca/big-data/companies?page='+str(i)
        else:
            url = 'https://themanifest.com/ca/big-data/companies'
        # Get request    
        result = requests.get(url)

        # Get the content
        content = result.content

        # Create soup object
        soup = BeautifulSoup(content,'lxml')

        # Get all companies list
        company_section = soup.find_all('li',{'class':'provider-card'})


        # Loop in the elements to scrape values
        for i in company_section:

            # url
            url = i.find('a',{'class':'track-website-visit'}).get("href")

            # Company Name
            company_name = i.find('h3',{'class':'provider-card__title'}).find('a').text.strip()

            # city
            city = i.find('span',{'class':'locality'}).text.strip()

            # card_details
            card_details = i.find('ul',{'class':'provider-card__details'}).find_all('span')

            # number_of_employees
            number_of_employees = card_details[1].text.replace('employees','')
            
            # projects_pudget
            projects_pudget = card_details[0].text
            if projects_pudget == 'Undisclosed':
                projects_pudget = projects_pudget
            else:
                projects_pudget = re.findall('[0-9]+,+[0-9]+',projects_pudget)[0]



            print(url)
            print("*"*50)

            print(company_name)
            print("*"*50)

            print(city)
            print("*"*50)

            print(number_of_employees)
            print("*"*50)

            print(projects_pudget)
            print("*"*50)
            time.sleep(3)
            # write in CSV file
            wr.writerow([company_name, city, number_of_employees, projects_pudget,url])

```
